### Importando as bibliotecas

```python
# Permite a geração de gráficos
import matplotlib.pyplot as plt

# Permite trabalhar com matrizes e vetores
import numpy as np 
```

### Primerio Gráfico com Matplotlib

```python
# Criando os vetores (conjunto de dados)
x = np.array([1,2,3,4,5,])
y = np.array([2,4,6,8,10])

# Plotar o Gráfico
plt.plot(x,y)
```

### Modificando os tipos de marcadores no gráfico

```python
plt.plot(x, y, marker = '*')
```

### Altrando as cores da linha

```python
plt.plot(x,y, marker = '.', c = 'r')
```

```python
plt.plot(
    x,y,            # Dados do gráfico
    marker = '*',    # Tipo de marcador
    c = 'green',     # Cor da linha
    ds = 'steps',   # A forma com que os pontos são conectados
    ls = '--',       # Estilo da linha
    mec = 'blue'    # Cor do marcador
)
```

### Matplotlib – adicionando outros elementos

Podemos ainda adicionar labels aos eixos, título ao gráfico, modificar o tamanho da fonte, dentre outras
possibilidades. Tome cuidado para que seu gráfico não tenha muitos elementos e acabe tirando a
atenção do que realmente importa.

```python
# Adiciona dados ao gráfico
plt.plot(x,y, c= 'r')

# Adicionar título ao eixo X
plt.xlabel('Valores de X', fontsize = '12')
plt.ylabel('Valores de Y', fontsize = '12')

# Adicionar título ao gráfico
plt.title('Título do Gráfico', fontsize = '20', c = 'purple')

# Altera as cores dos valores dos eixos
plt.xticks(fontsize = '8', c = 'blue')
plt.yticks(fontsize = '8', c = 'g')
```

### Múltiplas curvas no mecanismo gráfico

```python
import matplotlib.pyplot as plt
import numpy as np

# Retorna (Linspace) valores igualemte espaçados em um intervalo
x = np.linspace (0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)
plt.plot (x,y)
plt.plot (x,z)
```

### Matplotlib – Múltiplos subgráficos

```python
# Comando plt.suplot permite criar grupo de gráficos
# plt.subplot(linhas, colunas, posição elemento)

plt.subplot(2,1,1)  # Primerio gráfico
plt.plot(x,y)       # Conjunto de dados do primeiro gráfico
plt.title('Subplots')

plt.subplot(2,1,2) # Segundo gráfico
plt.plot (x,z)     # Conjunto de dados do segundo gráfico
```

### Gráficos na Vertical

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Subplots')
plt.subplot(1,2,2)
plt.plot(x,z)
```

# Seaborn – Gráficos mais elaborados

Seaborn é uma biblioteca baseada em Matplolib que traz uma interface mais atraente para os gráficos, além de facilitar a modificação de parâmetros default do Matplotlib e permite a utilização de dataframes.

```python
# O ALIAS (apelido) da biblioteca Seaborn fica como sns
import seaborn as sns
import numpy as np

# Retorna (Linspace) valores igualemte espaçados em um intervalo
x = np.linspace (0,2*np.pi,100)
y = np.sin(x)
```

### Seaborn – Datasets embutidos
O Seaborn possui diversos dados embutidos que podem ser utilizados para estudar a própria
biblioteca. Para sabermos quais os datasets estão disponíveis, podemos utilizar a função
get_datasets_names ().

```python
sns.get_dataset_names()
```

##### Criando um Data Frame com o dataset íris

```python
iris = sns.load_dataset('iris')
iris.head()
```

##### Seaborn – Ajustando os gráficos

```python
import matplotlib.pyplot as plt
sns.set()
plt.plot (x,y)
```

```python
# Alçterando tema para claro
sns.set_style('whitegrid')
plt.plot(x,y)
```

```python
# Alçterando tema para escuro
sns.set_style('dark')
plt.plot(x,y)
```

##### Seaborn – Dados categóricos

```python
# Gráfico de barras
sns.barplot(x='species',y='petal_length',data=iris)
```
