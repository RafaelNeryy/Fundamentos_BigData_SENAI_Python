```python
# Importar a biblioteca 
import pandas as pd

# Declarar as variáveis para acessar o arquivo
path = 'C:\\Users\\Administrador\\Documents\\Modelos Personalizados do Office\\Rafael Nery - Fundamentos de Big Data e Data Analytics com Python\\003 - Arquivos\\'
file = 'dados-vendas.xlsx'

# Acessar os dados do arquivo e colocar em um Data Frame
dados = pd.read_excel(path + file, sheet_name='Vendas')
```

7. 
Crie um dataframe chamado df_vendedores, para atribuir valores a ele você deve selecionar apenas a coluna Vendedores e remover as duplicadas do mesmo. Confirme se realmente o comando removeu as duplicadas. 

Após isso, crie uma lista com o seguinte nome listaVendedores, essa lista deve receber os dados do dataframe df_vendedores. Para isso, você deve converter o dataframe em lista, utilize o comando dataframe.tolist().

```python
# Seleciona apenas a coluna vendedor e remove as duplicatas 
df_vendedores = dados['Vendedor'].drop_duplicates()

# Converter o Pandas SERIES em uma lista 
lista_vendedores = df_vendedores.tolist()

lista_vendedores
```

8. 
Construa uma estrutura de repetição (FOR) para que a variável vendedor receba os nomes que estão na lista listaVendedores e filtre o dataframe dados, utilizando a coluna Vendedores para que gere um novo dataframe (df_final) para cada vendedor. Ainda na estrutura de repetição, você deve inserir uma linha de código para gerar um arquivo em Excel para cada vendedor, utilize o comando dataframe.to_excel. Você deve converter o dataframe em lista, utilize o comando dataframe.tolist().

```python
# Estrutura para gerar os arquivos de forma automática
for vendedores in lista_vendedores:

    df_final = dados[dados['Vendedor'] == vendedores]
    path_new = path + '...Relatório Vendedores\\'

    # Salvar o arquivo na pasta
    df_final.to_excel(path_new + vendedores + '.xlsx', index= False)
```

```python
# Categoria 
df_categoria = dados['Categoria'].drop_duplicates()

lista_categoria = df_categoria.tolist()

for categoria in lista_categoria:
    df_final = dados[dados['Categoria'] == categoria ]
    path_neew = path + '...Relatório Categoria\\'
    df_final.to_excel(path_neew + categoria + '.xlsx', index= False)
```

```python
# Segmento
df_seg = dados['Segmento'].drop_duplicates()

list_seg = df_seg.tolist()

for segmento in list_seg:
    df_final = dados[dados['Segmento'] == segmento]
    path_neww = path + '...Relatório Segmentos\\'
    df_final.to_excel(path_neww + segmento + '.xlsx', index= False)
```

```python
# Fabricante
df_fab = dados['Fabricante'].drop_duplicates()
list_fab = df_fab.tolist()

for fabricante in list_fab:
    df_final = dados[dados['Fabricante'] == fabricante]
    pathnew = path + '...Relatório Fabricante\\'
    df_final.to_excel(pathnew + fabricante + '.xlsx', index= False)
```

```python
# Cidade
df_city = dados['Cidade'].drop_duplicates()
list_city = df_city.tolist()

for cidade in list_city:
    df_final = dados[dados['Cidade'] == cidade]
    paathnew = path + '...Relatório Cidade\\'
    df_final.to_excel(paathnew + cidade + '.xlsx', index= False)
```

```python
# Estado
df_est = dados['Estado'].drop_duplicates()
list_est = df_est.tolist()

for estado in list_est:
    df_final = dados[dados['Estado'] == estado]
    pathhnew = path + '...Relatório Estado\\'
    df_final.to_excel(pathhnew + estado + '.xlsx', index= False)
```
