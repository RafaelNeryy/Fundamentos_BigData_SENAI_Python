##### 2. Qual a cobertura de clientes por ano? 
(Cobertura de clientes = total de clientes que efetuaram uma compra) 

```python
# Bloco utilizado para combinar as colunas Ano, Mês, Dia e MesNum do Data Frame dfData,
# para que seja possível agregar os dados de modo a visualizar as informações por ano.

df_merge = file_combined.merge(dfData,                  # Tabela da direita (fornecedora das colunas faltantes)
                               left_on = 'DataEmissao', # Coluna que será utilizada com valor procurado na tabela da esquerda.
                               right_on = 'Data',       # Coluna da tabela da direita que será utilizada para busca
                               how = 'left')            # É a forma que os dados serão inseridos, ESQUERDA <---- DIREITA

df_merge 
```

```python
# Cria um novo Data Frame, considerando apenas as colunas ANO e cdCliente do file_combined
df_contagem_cliente = df_merge.loc[:,['Ano', 'cdCliente']]

# Remover a duplicata para calcular a cobertura de clientes. Sendo que para este cálculo é necessário eliminar
# os registros duplicados, considera todas as colunas
df_contagem_cliente = df_contagem_cliente.drop_duplicates()

# Agrupa os dados, por meio da contagem, separando os dados através da coluna ANO
df_contagem_cliente = df_contagem_cliente.groupby(['Ano']).count()

df_contagem_cliente
```

```python
sns.barplot(x=df_contagem_cliente.index, y='cdCliente', data= df_contagem_cliente, orient= 'v')
```
