##### Exercício 01 - Escreva um programa Python para desenhar uma linha com rótulo adequado no eixo x, eixo y e um título.

```python
# Importar a bilioteca
import matplotlib.pyplot as plt

# Gerar os dados para inserir no gráfico
x = range(1,50)
y = [value * 3 for value in x]

print('Valores de X')
print(range(1,50))
print('Valores de Y (Três vezesX)')
print(y)

plt.plot(x,y, marker = '.')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Draw a line')
```

##### Exercício 02 - Escreva um programa Python para desenhar uma linha usando determinados valores de eixo com rótulo adequado no eixo x, eixo y e um título.

```python
import numpy as np

x = np.array([1,2,3])
y = np.array([2,4,1])

plt.plot(x,y, marker = '.', c = 'r')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Sample Graph!!!1')
```

```python
# Importar bibliotecas

import matplotlib.pyplot as plt
import pandas as pd

# Mapear o caminho do documento
path = 'C:\\Users\\Administrador\\Documents\\Modelos Personalizados do Office\\Rafael Nery - Fundamentos de Big Data e Data Analytics com Python\\003 - Arquivos\\'
file = 'fdata.csv'

# Abrir documento
# O argumento index_col permite selecionar com qual coluna será índice Data Frame. 
# Exemplo: 0 - Primeira coluna como índice 
# Segunda coluna como índice e none - padrão
df = pd.read_csv(path + file, sep = ',', index_col = None)

# Visualizar o cabeçalho
df.head()

# Plotar o gráfico
df.plot()
```
