### Exercício 18
Faça um programa que pergunte o preço de três produtos e informe qual
produto você deve comprar, sabendo que a decisão é sempre pelo mais barato.

```python
p1 = float(input('Insira o valor do 1º PRODUTO: R$ '))
p2 = float(input('Insira o valor do 2º PRODUTO: R$ '))
p3 = float(input('Insira o valor do 3º PRODUTO: R$ '))

if p1 < p2 and p1 < p3:
    print(f'Este produto é mais barato: Valor - R$ {p1}')
          
elif p2 < p1 and p2 < p3:
    print(f'Este produto é mais barato: Valor - R$ {p2}')
          
elif p3 < p1 and p3 < p2:
    print(f'Este produto é mais barato: Valor - R$ {p3}')
          
```

### Exercício 21

• salários até R$ 280,00 (incluindo) : aumento de 20%

• salários entre R$ 280,00 e R$ 700,00 : aumento de 15%

• salários entre R$ 700,00 e R$ 1500,00 : aumento de 10%

• salários de R$ 1500,00 em diante : aumento de 5%


Após o aumento ser realizado, informe na tela:


• salário antes do reajuste;

• percentual de aumento aplicado;

• valor do aumento;

• novo salário, após o aumento.

```python
salario = float(input('Insira o seu salário: R$ '))

if salario <= 280:
    print(f'Salário antes do reajuste: {salario}; Percentual de aumento aplicado: {0.2*100}; Valor do aumento: {salario*0.2}; Novo salário: {(salario+salario*0.2)}.')
    
elif salario <= 700:
    print(f'Salário antes do reajuste: {salario}; Percentual de aumento aplicado: {0.15*100}; Valor do aumento: {salario*0.15}; Novo salário: {(salario+salario*0.15)}.')
    
elif salario <= 1500:
    print(f'Salário antes do reajuste: {salario}; Percentual de aumento aplicado: {0.1*100}; Valor do aumento: {salario*0.1}; Novo salário: {(salario+salario*0.1)}.')
    
else:
    print(f'Salário antes do reajuste: {salario}; Percentual de aumento aplicado: {0.05*100}; Valor do aumento: {salario*0.05}; Novo salário: {(salario+salario*0.05)}.')
    
```
