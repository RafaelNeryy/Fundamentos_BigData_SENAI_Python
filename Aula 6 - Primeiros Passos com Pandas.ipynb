## Seção 1 - Trabalhando com PANDAS SERIES
Uma série PANDAS é como uma columa de uma tabela, ou seja, uma matriz de apenas uma coluna

```python
#Utilizando o comando pd.Series()

idades = pd.Series([15,20,25,30,40,50,60])
print(idades)
```

```python
#Retornar apenas os valores da Série

idades.values
```

```python
#Retornar apenas os índices dos registros 

idades.index
```

```python
#Alterando os índices

indice_novo = ['Ana','Amanda','Jacqueline','Larissa','Roberto','João','Alan']

idades_list = [15,20,25,30,40,50,60]

nova_serie = pd.Series(idades_list,index= indice_novo)

print(nova_serie)
```

### Acessando informação de uma Série

```python
print(nova_serie['João'])
print(nova_serie['Ana'])
```

## Seção 2 - Acessando DataFrame

Pandas Data Frame é uma estrutura de dados bidimensional, como uma matriz com múltiplas linhas e colunas. 

```python
import pandas as pd

dados = {
    'calorias': [420,380,390],
    'duracao': [50,40,45]
}

#Carregando dados para um DataFrame
df = pd.DataFrame(dados)
print(df)

#Selecionando a coluna "calorias"
print(df['calorias'])

#Selecionando a coluna "calorias"
print(df['duracao'])
```

```python
#Verificando apenas os valores do DataFrame

df.values
```

```python
#Verificando os índices do Data Frame

df.index
```

### Estatísticas Básicas

A função "df.describe()" exibe as estatísticas básicas dos dados numéricos.


```python
df.describe()
```

### Estatíticas de Dados Categóricos (não numéricos)

Para estatíticas de variáveis categóricas, precisamos adicionar o argumento  "include = ['O']"

TOP -> O valor que mais aparece nos campos

FREQ -> Frequência (quantidade de vezes) que o valor mais comum aparece (TOP)

COUNT -> Quantidade de elementos existentes nos campos

UNIQUE -> Quantidade de elementos únicos (sem repetição) existentes nos campos

```python
df.describe(include=['O'])
```

### Estatística básica de todos os dados

Ainda é possível aplicar o "pd.describe" a todos os dados. Para isso, precisamos adicionar o argumento "include = 'all'"

```python
df.describe(include = 'all')
```
