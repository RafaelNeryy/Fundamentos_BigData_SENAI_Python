#### Criando um gráfico de colunas

```python
graficos = sns.barplot(x='Categoria', y='ValorVenda', data=df, orient='v')
# O método sns.barplot vair retornar uma lista de sub métodos utilizando os catainers para acessar o texto dos rótulos para cada coluna, 
# desta forma o python entende que ele deve adicionar um rótulo em cada coluna, de acordo com o comando graficos.conta

graficos.bar_label(graficos.containers[0])
```

#### Gráfico de Linha - Data x ValorVenda


```python
# Cria um novo dataframe utilizando as colunas Data Venda e ValorVenda
# e classificando o data frame com base na coluna Data Venda
base = df[['Data Venda','ValorVenda']].sort_values('Data Venda')

# Agrupa os dados por meio do campo Data e realiza a soma
base = base.groupby('Data Venda').sum()
```

```python
y = base['ValorVenda']
x = base.index

plt.plot(x,y)
```
